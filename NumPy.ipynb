```python
#sigmoid
  #It is mostly used in ml algorithm name logistic Regression and deep learning
def sigmoid(array):
  return 1/(1+np.exp(-(array)))
a=np.arange(10)
sigmoid(a)
#here sigmoid is not a built-in function we solve it by built-in function name 'exp(a)'
```

```python
# mean squared error

 #Mean Squared Error measures the average squared difference between actual values and predicted values.
 #square((actual value − predicted value))
# What MSE tells you
#   Lower MSE → better model
#   Higher MSE → worse prediction
#   MSE is always ≥ 0
#   Perfect model → MSE = 0
#uses
  #Strong penalty for big errors
  #Differentiable → works with gradient descent
  #Standard loss function for regression
```

```python
# mean squared error
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)
print(actual)
print(predicted)
```

```python
def mse(actual,predicted):
   return np.mean((actual-predicted)**2)

mse(actual,predicted)
```

```python
#binary cross entropy
  #Binary Cross Entropy measures how well predicted probabilities match actual binary outcomes (0 or 1).

# Why log is used (IMPORTANT)
#   Converts probabilities into penalties
#   Makes loss smooth and differentiable
#   Enables gradient descent

# What BCE tells you
#   Lower BCE → better classifier
#   0 → perfect prediction
#   Very wrong predictions → huge loss
```

```python
#binary cross entropy
def bce(actual,predicted):
   predicted = np.clip(predicted, 1e-15, 1 - 1e-15)
   return (actual* (np.log(predicted) + (1-actual))* np.log(1-predicted))
bce=(actual,predicted)
```

```python
m=np.array([1,2,3,4,5,6,7])
n=np.array([5,7,9,11,8,2,10])
print(np.union1d(m,n))
print(np.intersect1d(m,n))
print(np.setdiff1d(m,n))
print(np.setdiff1d(n,m))
print(np.setxor1d(m,n))#remove common element from both
print(np.in1d(m,6))#to check the element
print(np.in1d(n,11))#to check the element
print(m[np.in1d(m,6)])#to check the element
```
