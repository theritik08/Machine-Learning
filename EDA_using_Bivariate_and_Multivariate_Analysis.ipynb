```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
titanic=pd.read_csv('train.csv')
titanic
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
iris=sns.load_dataset('iris')
iris
```

## **1. Scatterplot(Numerical-Numerical)**

```python
#This is Multivariate Analysis
sns.scatterplot(data=tips, x='total_bill', y='tip', hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.show()
```

# **2.Bar Plot(Numerical-Categorical)**

```python
sns.barplot(data=titanic, x='Pclass', y='Age',hue=titanic['Sex'])
```

# **3. Box Plot(Numerical-Categorical)**

```python
sns.boxplot(data=titanic,x='Sex',y='Age')
```

```python
q1=titanic['Age'].quantile(0.25)
q3=titanic['Age'].quantile(0.75)
print(f"Q1 is {q1} and Q3 is {q3}")
```

# **4.KDE Plot(Numerical-Categorical)**

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived')
```

# **5. HeatMap(Categorical-Categorical)**

```python
titanic.head(3)
```

```python
 #crosstab is a function used to compute a frequency table (contingency table) of two or more categorical variables
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
#now plot the graph for this
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
#.groupby() is a fundamental function in Pandas used to split data into groups, apply a function, and combine the results.
#People survived in evary classes
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
#we can also plot using bar
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

# **6. ClusterMap(Categorical-Categorical)**

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
#A dendrogram groups similar rows or columns together based on distance, and its branch heights show how distinct those groups are
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

# **7. Pairplot**

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
sns.pairplot(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
sns.pairplot(iris,hue='species')
```

# **8. Lineplot**
A line plot (or line chart) is primarily used to display trends and changes over a continuous interval, most commonly over time.

```python
flights.head()
```

```python
new=flights.groupby('year').sum(numeric_only=True).reset_index()
new
```

```python
sns.lineplot(x=new['year'], y=new['passengers'], marker="o")
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
