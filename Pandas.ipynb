**Pandas Series:**
It is like a column in a table.it is a 1-D array holding data of any type.
In 90% of cases object is treat like a string.

**Series from lists**

```python
#string
country=['India','Pakistan','USA','Nepal','srilanka']
pd.Series(country)
```

```python
#integers
runs=[13,24,25,45,38,28]
pd.Series(runs)
```

```python
#custom index
marks=[67,98,70,100]
subjects=['maths','english','science','sst']
pd.Series(marks,index=subjects)
print(pd.Series(marks,index=subjects,name='Ritik Marks'))
```

# **Creating DataFrame**

2-D structure of data containing rows and columns.

columns and rows are call as series and table as DataFrame

```python
 #using list
student_data=[
     [100,80,10],
     [90,70,7],
     [120,70,14],
     [80,50,2]
 ]
pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
#using dictionary
student_dict={
    'Name':['nitish','ankit','rupesh','rishabh','amit','ankita'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,100,70,50,0,0],
    'package':[10,7,14,2,0,0]
}
students=pd.DataFrame(student_dict)
students.set_index('Name',inplace=True)#the index has changed to Names when we use inplace=true
students
```

# **Filtering a DataFrame**

```python
ipl.head(2)
```

```python
#find all the final winners
mask=ipl['MatchNumber']=='Final'
new_df=ipl[mask]
new_df[['Season','WinningTeam']]
#same thing but using Fancy indexing in one line
ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
#how many over finishes have occured
ipl['SuperOver']=='Y'#you get the answer in boolean form
ipl[ipl['SuperOver']=='Y']#if you put this condition in your dataframe yo get full display
```

```python
ipl[ipl['SuperOver']=='Y'].shape[0]#To get how many matches have same condition
```

```python
#How many matches has csk won in kolkata
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=='Chennai Super Kings')]
```

```python
#Toss winner is Match Winner in percentage
ipl[ipl['TossWinner']==ipl['WinningTeam']]#give the full DataFrame
(ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl.shape[0])*100#To get the Percentage
```

```python
#movies with rating higher than 8 and votes greater than 10000
movies[(movies['imdb_rating'] >8) &(movies['imdb_votes']>10000)]
```

```python
#Action movies with rating higher than 7.5
#since genres has differnt data so first we seperate the Action data
movies['genres']#.str.split('|').apply(lambda x:'Action' in x)
#movies[(movies['genres']=='Action') &(movies['imdb_rating']>7.5)]
```
