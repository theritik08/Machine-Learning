# **API(Application Programming Interface)**
Application Programming Interface, is a set of rules that allows different software applications to communicate and exchange data with each other. It acts as a digital middleman, delivering a request from one system to another and bringing the response back

**TMDB API(TOP RATED MOVIES)**

```python
import pandas as pd
import requests
```

```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=fab97f3edabaf2d64d83d65ef5dc35e0&language=en-US&page=1')
```

* **<Response[200]>**:means Everything Fine
* **<Response[404]>**:data not available
* **<Response[500]>**:server down




```python
#To explicitly convert the data to JSON Formate
#when you write result you get particular secion result from json
response.json()['results']
```

```python
pd.DataFrame(response.json()['results'])
```

```python
#Now extract the important thing
temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','overview','popularity','vote_average','vote_count']]
temp_df
```

```python
df=pd.DataFrame()
df
```

```python
for i in range(1,501):
  response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=fab97f3edabaf2d64d83d65ef5dc35e0&language=en-US&page={}'.format(i))
  temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','overview','popularity','vote_average','vote_count']]
  df = pd.concat([df, temp_df], ignore_index=True)
```
