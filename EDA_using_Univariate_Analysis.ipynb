```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/train.csv")
df
```

# **1.Categorical Data**

a. **Countplot**:

Use to show the raw frequency counts of categorical data (e.g., counting passengers per class).

```python
sns.countplot(x='Sex', data=df, legend=False)
plt.show()
```

b.**PieChart**:

Use to show the percentage breakdown of a single category relative to the whole (e.g., proportion of genders).

```python
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%')
```

# **2. Numerical Data**

a. **Histogram**:

Use to see the shape, spread, and frequency distribution of continuous numerical data (e.g., identifying common age brackets).

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=50)
```

b. **Distplot**:

Use to see the shape, spread, and frequency distribution of continuous numerical data (e.g., identifying common age brackets).

```python
#It is removed in new seaborn version inplace we use histplot
sns.distplot(df['Age'])
```

c. **Histplot**

```python
sns.histplot(data=df, x='Age', bins=20, kde=False)
plt.show()
```

d. **Boxplot**:

Use to compare numerical data distributions across categories and spot outliers (e.g., comparing ticket fares across survival groups).

```python
sns.boxplot(data=df, x='Age')
plt.show()
```
